# Vector libraries

## Libraries

```python
from pygieons import Ecosystem

# Initialize
e = Ecosystem(plot_type="vector", log=False)

# Prepare the table and plot it
e.prepare_table().show()
```

## Linkages

```python
# Initialize
e = Ecosystem(plot_type="vector+generic", log=False)

# Prepare the network and plot it
net = e.prepare_net()
net.show()
```
